```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the model
sunlight = 1000 # Amount of sunlight energy (W/m²)
algae_growth_rate = 0.2 # Proportion of sunlight energy used by algae for growth
zooplankton_grazing_rate = 0.1 # Proportion of algae energy consumed by zooplankton
zooplankton_mortality_rate = 0.05 # Proportion of zooplankton that die per unit time

# Initialize the variables of the model
algae_biomass = 100 # Amount of algae biomass (g/m²)
zooplankton_biomass = 10 # Amount of zooplankton biomass (g/m²)

# Simulate the model for 100 time steps
time = np.arange(100)
algae_biomass_history = []
for t in time:

    # Calculate the amount of sunlight energy absorbed by algae
    algae_sunlight_energy = algae_biomass * sunlight * algae_growth_rate

    # Calculate the amount of algae energy consumed by zooplankton
    zooplankton_grazing_energy = zooplankton_biomass * algae_biomass * zooplankton_grazing_rate

    # Calculate the net growth rate of algae
    algae_net_growth_rate = algae_sunlight_energy - zooplankton_grazing_energy

    # Calculate the net mortality rate of zooplankton
    zooplankton_net_mortality_rate = -zooplankton_mortality_rate * zooplankton_biomass

    # Update the algae and zooplankton biomass
    algae_biomass += algae_net_growth_rate
    zooplankton_biomass += zooplankton_grazing_energy - zooplankton_net_mortality_rate

    # Store the algae biomass at each time step
    algae_biomass_history.append(algae_biomass)

# Plot the algae biomass at all time steps
plt.plot(time, algae_biomass_history, label='Algae')

plt.xlabel('Time (steps)')
plt.ylabel('Biomass (g/m²)')
plt.legend()
plt.show()


```
